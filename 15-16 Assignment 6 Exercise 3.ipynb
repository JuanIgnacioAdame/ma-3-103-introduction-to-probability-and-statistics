### Exercise 3.1

Assume that match lengths are normally distributed. Construct a $95\%$ confidence interval for the standard deviation of match lengths.

Using Fact 21.6.1 from Lecture 21, we have that
\begin{equation}
\frac{(n-1)S^2}{\sigma^2}\sim\text{chi-square}(n-1).
\end{equation}

Hence, a $(1-\alpha)$ confidence interval for $\sigma^2$ is given by
\begin{equation}
\frac{(n-1)s^2}{\chi^2_{\frac{\alpha}{2}}}<\sigma^2<\frac{(n-1)s^2}{\chi^2_{1-\frac{\alpha}{2}}}.
\end{equation}

Note that this confidence interval is not unique.  Below we show a few examples of a couple different $1-\alpha$ confidence intervals. $\square$

```python
import numpy as np
```

```python
# raw data
match_lengths = np.array([73, 76, 59, 104, 114, 106, 79, 74, 142, 129, 95, 56, 84, 142, 106, 75])
```

```python
# number of observations
n = match_lengths.size
# unbiased population sample standard deviaton
S2 = np.var(match_lengths)*n/(n-1)
```

```python
from scipy.stats import chi2
```

```python
X = chi2(df=n-1)
```

```python
# different confidence intervals
# notice that some choices are shorter!
for a in [0.04, 0.041, 0.042]:
    I = [np.sqrt((n-1)*S2/X.ppf(a+0.95)), np.sqrt((n-1)*S2/X.ppf(a))]
    print(I)
    print(I[1]-I[0])
```

```python
from matplotlib import pyplot as plt
```

```python
plt.clf()

# x axis
plt.xlabel("Match label")

# y axis
plt.ylabel("Match length in minutes")

u = np.mean(match_lengths)
plt.scatter(range(len(match_lengths)), list(match_lengths))
plt.scatter([16], [u])
plt.scatter([17], [u])
plt.errorbar([16], [u], yerr=[[I[0]], [I[0]]], color='orange', label=r"$\mu\pm a$")
plt.errorbar([17], [u], yerr=[[I[1]], [I[1]]], color='green', label=r"$\mu\pm b$")
# legend
plt.legend(loc='upper left')
plt.show()
```
